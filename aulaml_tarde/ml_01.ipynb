```python
## Papagaio

# Features [1 sim, 0 não]
# Tem bico?
# Risco de extinção?
# Faz cocoricó?

papagaio1 = [1,1,0]
papagaio2 = [1,1,0]
papagaio3 = [1,1,1,]
```

```python
## Galinha

# Features [1 sim, 0 não]
# Tem bico?
# Risco de extinção?
# Faz cocoricó?

galinha1 = [1,0,1]
galinha2 = [1,0,1]
galinha3 = [1,1,1]
```

```python
dados = [papagaio1, papagaio2, papagaio3, galinha1, galinha2, galinha3]
```

```python
classes = [1,1,1,0,0,0]
```

```python
dados
```

```python
! pip install scikit-learn
```

```python
import sklearn
```

```python
from sklearn.svm import LinearSVC
modelo = LinearSVC()
```

```python
modelo.fit(dados, classes)
```

```python
# Features [1 sim, 0 não]
# Tem bico?
# Risco de extinção?
# Faz cocoricó?

# Tem bico, está em extinção e não faz cococricó

animal_misterioso = [1, 1, 0]
previsao = modelo.predict([animal_misterioso])
```

```python
previsao
```

```python
# Features [1 sim, 0 não]
# Tem bico?
# Risco de extinção?
# Faz cocoricó?

# Tem bico, está em extinção e não faz cococricó

animal_misterioso2 = [1, 0, 1]
previsao2 = modelo.predict([animal_misterioso2])
```

```python
previsao2
```

```python
# Dados de treino X 

treino_x = [papagaio1, papagaio2, papagaio3, galinha1, galinha2, galinha3]

# Rótulos (classes) treino Y

treino_y = [1, 1, 1, 0, 0, 0]
```

```python
# Animais misteriosos - para teste

misterio1 = [1, 1, 1]
misterio2 = [1, 1, 0]
misterio3 = [0, 1, 1]
```

```python
# Previsóes - dados de testes

teste_x = [misterio1, misterio2, misterio3]
previsoes = modelo.predict(teste_x)
```

```python
# Rótulos (classe) de teste avaliação - aqui é a sua provinha

teste_y = [0, 1, 1]
```

```python
# Importando módulo que já traz a função de acurácia encapsulada
from sklearn.metrics import accuracy_score

# Definindo uma variável para receber o retorno da função acurácia * 100 para porcentagem
taxa_de_acerto = accuracy_score(teste_y, previsoes) * 100

# Mostrando a acurácia
print(f"Acurácia: {taxa_de_acerto:.2f}%")
```
